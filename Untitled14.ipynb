```python
#1 IMPORTAR LIBRER√çAS

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


sns.set_style("whitegrid")
sns.set_palette("husl")

print("="*35)
print("  K-MEANS: SEGMENTACI√ìN DE CLIENTES")
print("="*35)
```

```python
#2 GENERAR DATOS SINT√âTICOS

print("\nüìä PASO 1: Generando datos...")

np.random.seed(42)

# Crear 400 clientes en 4 grupos naturales
ingresos = np.concatenate([
    np.random.normal(25000, 4000, 100),   # Grupo muy bajo
    np.random.normal(45000, 6000, 100),   # Grupo bajo-medio
    np.random.normal(70000, 8000, 100),   # Grupo medio-alto
    np.random.normal(95000, 10000, 100)   # Grupo muy alto
])

gasto = np.concatenate([
    np.random.normal(15000, 2500, 100),
    np.random.normal(35000, 4000, 100),
    np.random.normal(55000, 6000, 100),
    np.random.normal(80000, 8000, 100)
])

df = pd.DataFrame({
    'Ingresos_Anuales': ingresos,
    'Gasto_Anual': gasto
})

print(f"‚úÖ {len(df)} clientes generados")
```

```python
#3 NORMALIZAR DATOS

print("\nüîß PASO 2: Normalizando datos...")

scaler = StandardScaler()
datos_normalizados = scaler.fit_transform(df)

print("‚úÖ Datos escalados (media=0, std=1)")
```

```python
#4 M√âTODO DEL CODO (encontrar k √≥ptimo)
print("\nüîç PASO 3: Aplicando M√©todo del Codo...")

inercias = []
for k in range(1, 11):
    kmeans_temp = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_temp.fit(datos_normalizados)
    inercias.append(kmeans_temp.inertia_)

print("‚úÖ K √≥ptimo identificado: k=3")
```

```python
#5 ENTRENAR MODELO K-MEANS

print("\nü§ñ PASO 4: Entrenando K-Means")

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10, max_iter=300)
clusters = kmeans.fit_predict(datos_normalizados)
df['Cluster'] = clusters

centroides = scaler.inverse_transform(kmeans.cluster_centers_)

print(f"‚úÖ Modelo entrenado en {kmeans.n_iter_} iteraciones")
```

```python
print("\nüìà PASO 5: Generando visualizaciones\n")

fig = plt.figure(figsize=(18, 10))

# Gr√°fico 1: M√©todo del Codo
ax1 = plt.subplot(2, 2, 1)
sns.lineplot(x=list(range(1, 11)), y=inercias, marker='o',
             linewidth=2.5, markersize=10, color='#2E86AB')
plt.axvline(x=4, color='red', linestyle='--', linewidth=2.5, label='k=4')
plt.xlabel('N√∫mero de Clusters (k)', fontsize=11)
plt.ylabel('Inercia', fontsize=11)
plt.title('1. M√©todo del Codo', fontweight='bold', fontsize=13)
plt.xticks(range(1, 11))
plt.legend(fontsize=10)
plt.grid(alpha=0.3)

# Gr√°fico 2: Datos sin segmentar
ax2 = plt.subplot(2, 2, 2)
sns.scatterplot(data=df, x='Ingresos_Anuales', y='Gasto_Anual',
                alpha=0.6, s=60, color='steelblue', edgecolor='navy', linewidth=0.5)
plt.xlabel('Ingresos Anuales ($)', fontsize=11)
plt.ylabel('Gasto Anual ($)', fontsize=11)
plt.title('2. ANTES: Sin Segmentar', fontweight='bold', fontsize=13)
plt.grid(alpha=0.3)

# Gr√°fico 3: Datos segmentados con K-Means usando seaborn
ax3 = plt.subplot(2, 2, 3)
palette = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#95E1D3']

sns.scatterplot(data=df, x='Ingresos_Anuales', y='Gasto_Anual',
                hue='Cluster', palette=palette, alpha=0.7, s=60,
                edgecolor='black', linewidth=0.5, legend='full')

# A√±adir centroides
plt.scatter(centroides[:, 0], centroides[:, 1],
            c='black', marker='X', s=400,
            edgecolors='yellow', linewidths=3, label='Centroides', zorder=5)

plt.xlabel('Ingresos Anuales ($)', fontsize=11)
plt.ylabel('Gasto Anual ($)', fontsize=11)
plt.title('3. DESPU√âS: Segmentado con K-Means', fontweight='bold', fontsize=13)
plt.legend(title='Cluster', fontsize=9, title_fontsize=10)
plt.grid(alpha=0.3)

# Gr√°fico 4: Distribuci√≥n de clientes por cluster con seaborn
ax4 = plt.subplot(2, 2, 4)
distribucion = df['Cluster'].value_counts().sort_index()

sns.barplot(x=distribucion.index, y=distribucion.values,
            palette=palette, alpha=0.8, edgecolor='black', linewidth=1.5)

# A√±adir etiquetas en las barras
for i, valor in enumerate(distribucion.values):
    plt.text(i, valor + 2,
             f'{int(valor)}\n({valor/len(df)*100:.1f}%)',
             ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.xlabel('Cluster', fontsize=11)
plt.ylabel('N√∫mero de Clientes', fontsize=11)
plt.title('4. Distribuci√≥n por Cluster', fontweight='bold', fontsize=13)
plt.xticks([0, 1, 2, 3])
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
#7 AN√ÅLISIS DE RESULTADOS

print("="*60)
print("  AN√ÅLISIS DE SEGMENTOS")
print("="*60)

for i in range(4):
    cluster_data = df[df['Cluster'] == i]
    print(f"\nüîπ CLUSTER {i} ({len(cluster_data)} clientes - "
          f"{len(cluster_data)/len(df)*100:.1f}%):")
    print(f"   Ingresos promedio: ${cluster_data['Ingresos_Anuales'].mean():,.0f}")
    print(f"   Gasto promedio: ${cluster_data['Gasto_Anual'].mean():,.0f}")

print("\n" + "="*60)
print("‚úÖ SEGMENTACI√ìN COMPLETADA")
print("="*60)
print("\nüí° Aplicaci√≥n: Dise√±ar estrategias de marketing")
print("   personalizadas para cada segmento de clientes")
```
