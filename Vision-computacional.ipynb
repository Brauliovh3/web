Imagenes en Blanco y negro

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
print("Librerias ok")
```

```python
imagen =  np.zeros((10,10,1),dtype=np.uint8)
plt.imshow(imagen,cmap='gray')
plt.show()
```

```python
#Manipular pixeles
imagen[0,0] = 150
imagen[4,5] = 10
imagen[0,9] = 200
plt.imshow(imagen,cmap='gray')
```

```python
for i in range(10):
        imagen[i,i] = 255
plt.imshow(imagen,cmap='gray')
plt.show()
```

## Imagenes a color

```python
imagen_color = 200*np.ones((10,10,3),dtype=np.uint8)
plt.imshow(imagen_color)
plt.show()
```

```python
# Manipular colores

# Extraer canales
R = imagen_color[:,:,0]
G = imagen_color[:,:,1]
B = imagen_color[:,:,2]

# Modificar colores

R[:,:]=17
G[:,:]=254
B[:,:]=255

imagen_color[:,:,0] = R
imagen_color[:,:,1] = G
imagen_color[:,:,2] = B

# Mostrar imagen
plt.imshow(imagen_color)
plt.show()
```

# LECTURA Y MANIPULACION DE IMAGENES

```python
#Cargar imagen
from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]

imagen2 = cv2.imread(file_name)
plt.imshow(imagen2)
plt.show()
```

```python
# Modificando modelo de color

imagen_rgb = cv2.cvtColor(imagen2,cv2.COLOR_BGR2RGB)
plt.imshow(imagen_rgb)
plt.show()
```

```python
file_name = list(uploaded.keys())[0]

# Leer la imagen
imagen2 = cv2.imread(file_name)

# Convertir a RGB
imagen2_rgb = cv2.cvtColor(imagen2, cv2.COLOR_BGR2RGB)

# Crear figura con 2 imgenes
plt.figure(figsize=(14, 6))

# Imagen original
plt.subplot(1, 2, 1)
plt.imshow(imagen2)
plt.title('Imagen BGR')
plt.axis('off')

# Imagen convertida a RGB
plt.subplot(1, 2, 2)
plt.imshow(imagen2_rgb)
plt.title('Imagen Real')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Extraer y mostrar canales

```python
#Extraccion de canales
R,G,B = cv2.split(imagen_rgb)

#Mostrar los canales
fig = plt.figure(figsize=(16,10))

#Canal rojo
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(R,cmap='gray')
ax1.set_title('Canal Rojo')
plt.axis('off')
#Canal verde
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(G,cmap='gray')
ax2.set_title('Canal Verde')
plt.axis('off')
# Canal azul
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(B,cmap='gray')
ax3.set_title('Canal Azul')
plt.axis('off')
# Unir canales
imagen_unida = cv2.merge((R,G,B))

# Mostrar imagen unida
ax4 = fig.add_subplot(2,2,4)
ax4.imshow(imagen_unida)
ax4.set_title('Imagen Unida')
plt.axis('off')

plt.show()
```
