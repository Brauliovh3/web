# APLICACIÓN PRÁCTICA DE ALGEBRA LINEAL

```python
# Importar Librerias
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt
```

```python
# Cargar dataset
caras = fetch_olivetti_faces()
X = caras.data # Imagenes
y = caras.target # Etiquetas
y

# Mostrar un ejemplo de las imágenes
plt.imshow(caras.images[333], cmap = 'gray')
plt.title("Ejemplo de rostro - persona {y [100]}")
plt.show()
```

```python
# Aplicaciones Análisis de componentes Principales PCA
pca = PCA(n_components = 100)
x_reducido = pca.fit_transform(X)
x_reconstruido = pca.inverse_transform(x_reducido)
```

```python
# Realizar la comparación

# Seleccionar una imagen de ejemplo
indice = 0
imagen_original = caras.images[indice]
imagen_reconstruida = x_reconstruido[indice].reshape(64, 64)

# Mostrar comparación
fig, axes = plt.subplots(1, 2, figsize = (8, 4))
axes[0].imshow(imagen_original, cmap = 'gray')
axes[1].imshow(imagen_reconstruida, cmap = 'gray')
```
